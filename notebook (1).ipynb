```python
df['age'] = 2024 - df['model_year'] # A new column 'age' is added to the DataFrame `df`, containing the calculated age of each vehicle in years. The age is calculated by subtracting the model year from the current year (2024).
```

```python
def age_category(x):
    if x<5: return '<5'
    elif  x>=5 and x<10: return '5-10'
    elif x>=10 and x<20: return '10-20'
    else: return '>20' # Define a function to categorize vehicle age into different age categories. This function categorizes the age of vehicles into different age groups based on predefined criteria.
```

```python
df['age_category'] = df['age'].apply(age_category) # A new column 'age_category' is added to the DataFrame `df`, containing the age category of each vehicle based on the categorization criteria defined in the age_category function.
```

```python
df.head() #printing the first 5 rows to see if age category was added 
```
